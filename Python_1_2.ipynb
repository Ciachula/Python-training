## tworzenie nowej kolumny/ funkcja apply do manipulacji danymi

#### df['nazwa_nowej_kolumny'] = obliczenia

#### df.apply( lambda x: ,axis=1)

<li>axis = 1 (działania wykonywanie na kolumnach)
<li>pod zmienną x mamy całego DataFrame'a
<li>po : możemy wykonać dowolne obliczenia różnych kolumnach
    
#### df['nazwa_kolumny'].apply( lambda x: )

<li>axis = 0 (opcja domyślna nie potrzeba jej pisać)(działania wykonywane na wierszach pojedynczej kolumny)
<li>pod zmienną x mamy wiersz kolumny podanej po df['']
<li>po : możemy wykonać dowolne obliczenia wierszu kolumny    

```python
#stworzenie przykładowego DataFrame
dff = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'count': [5, 7, 12, 21, 1],
    'rating': [4, 4, 3.5, 15, 5]
})
dff
```

```python
# dostęp do jednej kolumny i wykonywanie obliczeń na niej
dff['dzielenie'] = dff.rating.apply(lambda x: x/5)
dff
```

```python
# dostęp do jednej kolumny i wykonywanie obliczeń na niej
dff['czy_1'] = dff['count'].apply(lambda x: 1 if x==1 else 0)
dff
```

```python
# obliczenia między dwiema kolumnami
dff['rating_count'] = dff.apply(lambda x: x['rating']/x['count'],axis=1)
dff
dff['rating_count2'] = dff.apply(lambda x: x['rating'],axis=1)
dff
```

```python
# dodać nową kolumne 'bedrooms_rooms', która powstanie po podzieleniu total_bedrooms/total_rooms
df['bedrooms_rooms'] = df.apply(lambda x: x['total_bedrooms']/x['total_rooms']*100, axis=1)
df
```

```python
# dodać nową kolumnę 'is_population30' jeśli wartość population jest mniejsza równa 30
df['ispopulation30'] = df.apply(lambda x: 1 if x['population']<=30 else 0, axis=1)
df
df['ispopulation30'].value_counts()
```

### regular expressions 

Zestaw znaków i metaznaków opisujących wzorzec w tekście
- znak wyszukuje odpowiadającą mu litere
- metaznak wyszukuje zbiór liter, który określa np. \d wyszukuje wszystkie liczby od 0-9<br>
Metaznaki:<br>
() - grupuje razem wiele tokenów i tworzy grupę przechwytywania w celu wyodrębnienia podciągu (wszystkie znaki ze środka zostaną wyświetlone w wyniku)<br>
[] - wyszukuje wszystkie pojedyncze znaki zadeklarowane wewnątrz [], stosowane gdy chcemy wyszukać więcej niż dwa różne znaki lub metaznaki
<br>
\w - przechwytuje dowolny znak alfanumeryczny(A-Za-z0-9) i znak podkreślenia <br>
\d - przechwytuje wszystkie cyfry od 0-9<br>
W celu wyszukania większego wzorca w tekście niż tylko pojedynczych znaków powielamy wcześniejsze tokeny<br>
np. <br>
\d+ - wyszukuje jeden lub więcej kombinacji cyfr 0-9<br>
\d* - wyszukuje zero lub więcej kombinacji cyfr 0-9<br>

Interaktywne środowisko: https://regexr.com/<br><br>
Metody:<br>

re.match() - zwraca szukany wzorzec (musi zaczynać się od pierwszego znaku)<br>
re.findall() - zwraca wszystkie teksty wyszukane poprzez wyrażenie <br>

<br>
Dokumentacja wyszukanie poprzez re.\<nazwa_metody><br>
https://docs.python.org/3/library/re.html

```python
txt='8.3252$'
res = re.match('[\d.]+',txt)
float(res.group())
```
